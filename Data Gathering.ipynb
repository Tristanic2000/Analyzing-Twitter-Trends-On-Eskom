```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list = []

for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f"eskom since:2022-03-09 until:2022-07-16").get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.user.displayname, tweet.lang])
```

```python
tweets_df = pd.DataFrame(tweets_list, columns=["datetime", "tweet_id", "text", "username", "like_count", "display_name", "lang"])
```
