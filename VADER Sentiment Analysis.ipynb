```python
import numpy as np
import pandas as pd
import re
import string
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# nltk.download('vader_lexicon')
```

```python
tweets_df = pd.read_csv('data/tweets_cleaned.csv')
tweets_df.head()
```

```python
sia = SentimentIntensityAnalyzer()
tweets_df['score'] = tweets_df['text'].apply(sia.polarity_scores)
tweets_df['neg'] = tweets_df['score'].apply(lambda x: x['neg'])
tweets_df['neu'] = tweets_df['score'].apply(lambda x: x['neu'])
tweets_df['pos'] = tweets_df['score'].apply(lambda x: x['pos'])
tweets_df['compound'] = tweets_df['score'].apply(lambda x: x['compound'])
```

```python
def calculate_sentiment(compound):
    sentiment = 'Neutral'
    if compound >= 0.75:
        sentiment = 'Very Positive'
    elif compound >= 0.05:
        sentiment = 'Positive'
    elif compound <= -0.75:
        sentiment = 'Very Negative'
    elif compound <= -0.05:
        sentiment = 'Negative'
    
    return sentiment
```

```python
tweets_df['sentiment'] = tweets_df['compound'].apply(calculate_sentiment)
tweets_df.head()
```

```python
tweets_df.to_csv('data/twitter_sentiment_analysis.csv', index = False)
```
